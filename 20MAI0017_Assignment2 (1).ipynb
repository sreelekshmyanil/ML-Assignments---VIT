# 6. Display the ‘Employment’ of customers who mostly ‘Rent’

When you have more than one row that contains the minimum or maximum value in a column, the topmost row containing that value will be returned in the result.

Method 1 : Using group by

```python
#first group by householder then employement
rnt = loan.groupby(['householder','employment'])
```

```python
#count of distinct employment in each householder
rnt.count()
```

```python
tmp = rnt.count()
```

```python
type(tmp)
```

```python
#filter data only for rent
x = tmp.loc['RENT']
```

```python
x
```

```python
m = x['customer_id'].max()
print("The maximum count is  : ",m)
```

```python
print("The ‘Employment’ of customers who mostly ‘Rent’  : ", x['customer_id'].idxmax())
```

```python
#but we have more than one value
```

```python
type(x['customer_id'])
```

```python
#iterate through series
for index, value in x['customer_id'].items():
    if value == m:
        print("The ‘Employment’ of customers who mostly ‘Rent’  : ",index)
```

# Method 2 : Instead of grouping by two variables, we extract id and then process

```python
#group by householder classes
rnt = loan.groupby('householder').groups['RENT']
#ids of those are rent
```

```python
#getting employment of those who rent as list
l = []
for i in rnt:
    l.append(loan.iloc[i]['employment'])

print(l)
```

```python
# importing Counter module 
from collections import Counter 
```

```python
#calculate unique employment with frequency and sort in descending order
top = sorted(list(zip(Counter(l).keys(),Counter(l).values())),key = lambda x: x[1],reverse=True)[0]
```

```python
for i in zip(Counter(l).keys(),Counter(l).values()):
    if i[1] == top[1]:
        print("The ‘Employment’ of customers who mostly ‘Rent’  : ", i)
```
